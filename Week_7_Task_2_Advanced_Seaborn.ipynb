## 2️⃣ Distribution and Relationship Plots (40 Marks)

### Histogram/Distplot: Display the distribution of a numeric variable. (10 Marks)


```python
sns.histplot(df['latitude'])
```

**The histogram of latitude values will display how the latitudes of the stadiums are distributed geographically.This reveals most stadiums are concentrated in latitude 40 region.**

### Pairplot: Show pairwise relationships between numerical variables. (10 Marks)


```python
sns.pairplot(df[['latitude', 'longitude','stadium_capacity','joined'  ]])
```

### Heatmap: Visualize correlations between numerical features. (10 Marks)

```python
sns.heatmap(df[['latitude', 'longitude','stadium_capacity','joined'  ]].corr(),annot=True)
```

**Latitude and Longitude:** There is weak correlation between latitude and longitude, as expected. These two geographic coordinates are independent of each other; one controls north-south positioning, while the other controls east-west positioning.

**Stadium Capacity and Latitude/Longitude:** The heatmap shows weak correlation between stadium capacity and both latitude and longitude. This suggests that a stadium's geographic location (in terms of its latitude and longitude) doesn't have a strong linear relationship with its size. Factors such as population density, urban development, or sports culture are more likely to influence stadium capacity.

**Stadium Capacity and Joined:**The correlation between stadium capacity and the "joined" variable (the year the stadium was built or established) shows a moderate positive correlation. Newer stadiums might tend to have larger capacities due to modern trends in construction, a reflection of the trend towards larger, multi-purpose venues being built in recent years.




### Jointplot: Analyze the relationship between two variables, including marginal distributions. (10 Marks)


```python
sns.jointplot(x='stadium_capacity', y='latitude', data=df, kind='scatter', color='blue')
```

**The scatter plot shows the relationship between stadium capacity and latitude. Latitude represents the geographic location of each stadium, with values increasing as you move north or south from the equator.There doesn't seem to be a strong linear relationship between stadium capacity and latitude. Stadiums located at different latitudes (both in the northern and southern hemispheres) appear to have a variety of capacities.**

# 3️⃣ Categorical Data Visualization (20 Marks)

### Bar Plot: Compare categories using aggregation. (10 Marks)


```python
# Aggregate stadium capacity by state
state_capacity = df.groupby('state')['stadium_capacity'].sum().sort_values(ascending=False)

# Create the bar plot
plt.figure(figsize=(12, 8))
state_capacity.plot(kind='bar', color='skyblue')
plt.title('Total Stadium Capacity by State', fontsize=16)
plt.xlabel('State', fontsize=12)
plt.ylabel('Total Stadium Capacity', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

**States with major metropolitan areas or sports-focused regions seem to dominate in terms of total capacity. This is likely because larger cities tend to host more stadiums and larger venues. For instance, California, Texas, and Florida likely have more professional sports teams and larger venues, contributing to their higher stadium capacities. Whereas states like Colorado and Kansas have less stadium capacity.**

### Box Plot: Visualize the distribution of a numeric variable across categories. (10 Marks)

```python
# Creating the box plot
plt.figure(figsize=(12, 10))
sns.boxplot(x='state', y='stadium_capacity', data=df)
plt.xticks(rotation=90)  # Rotate state names for readability
plt.title('Distribution of Stadium Capacity Across States')
plt.xlabel('State')
plt.ylabel('Stadium Capacity')
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

**States like California,Florida,Ohio and Texas have a wide range of stadium capacities, suggesting a higher variation in the size of stadiums across teams.There is a noticeable trend where states with metropolitan areas (like California and Texas) tend to have higher stadium capacities. This could be due to larger populations and more demand for larger venues.Overall, the plot highlights the impact of factors like state population, local support for soccer, and geographical factors on the size of stadiums. Some outliers show that certain teams may have invested more in stadiums to cater to larger fanbases or because they are located in major metropolitan areas.**
