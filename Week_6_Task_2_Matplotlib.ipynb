## i. Line Plot: Show a trend or progression of a variable over time or another index. (10 Marks)


```python
df['date'].value_counts()
```

```python
match_counts = df['date'].value_counts().sort_index()
match_counts
```

```python
# Count the number of matches played in each season
season_counts = df['series_name'].value_counts().sort_index()

# Plotting the line plot for seasons
plt.figure(figsize=(10, 5))
plt.plot(season_counts.index, season_counts.values, marker='o', linestyle='-', color='g')

plt.title('Matches Played by Season', fontsize=14)
plt.xlabel('Season')
plt.ylabel('Number of Matches', fontsize=12)
plt.xticks(rotation=90)
plt.grid(True)

plt.show()
```

### Insights
 **Some seasons have more matches, which could be due to expanded league formats, inclusion of more teams, or changes in the overall structure of the competition.You can observe how the number of matches fluctuated over different seasons.**

## ii. Bar Chart: Compare categories or groups. (10 Marks)

```python
#Group the data by 'home_team_name' and count the number of matches for each team
team_match_counts = df['home_team_name'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(12, 6))  # Set the figure size
team_match_counts.plot(kind='bar', color='skyblue')

# Customizing the plot
plt.title('Number of Matches Played by Each Team', fontsize=14)  # Title of the plot
plt.xlabel('Team', fontsize=12)  # Label for X-axis
plt.ylabel('Number of Matches', fontsize=12)  # Label for Y-axis
plt.xticks(rotation=90)  # Rotate the x-axis labels for better readability
plt.grid(True, axis='y')  # Add grid lines on the y-axis

# Display the plot
plt.show()
```

### Insights
**The chart helps us see which teams have participated the most in the league. Teams with higher bars have played more matches. While the bar chart does not show win/loss data, it gives an overall idea of how much exposure each team has had in the league. Teams with more matches could have been more successful in advancing through the stages.**

##iii. Scatter Plot: Illustrate relationships between two numeric variables. (10 Marks)

```python
# Extract match scores from the 'result' column (assume match result format: "Team1 beat Team2 (Score1 - Score2)")
df['match_score'] = df['result'].apply(lambda x: x.split('(')[1].split(' - ')[0] if isinstance(x, str) and '(' in x else None)
df['match_score'] = pd.to_numeric(df['match_score'], errors='coerce')  # Convert to numeric, ignoring errors

# Convert 'toss_winner' to numeric team ID (as an example, assuming 'toss_winner' contains team IDs)
df['toss_winner'] = pd.to_numeric(df['toss_winner'], errors='coerce')  # Convert to numeric values

# Drop rows with missing values (if any)
df_clean = df.dropna(subset=['match_score', 'toss_winner'])

# Plotting the scatter plot
plt.figure(figsize=(10, 6))  # Set figure size
plt.scatter(df_clean['match_score'], df_clean['toss_winner'], color='red', alpha=0.5)

# Customizing the plot
plt.title('Scatter Plot: Match Score vs Toss Winner Team ID', fontsize=14)
plt.xlabel('Match Score', fontsize=12)
plt.ylabel('Toss Winner Team ID', fontsize=12)
plt.grid(True)  # Adding grid lines for better readability

# Display the plot
plt.show()
```

### Insights
**The match scores vary across a wide range, indicating that the Pro Kabaddi League has had matches with significant score differences. The plot might not show a direct relationship with the toss winner, suggesting that other factors (team performance, strategy, etc.) play a more significant role in determining the match outcomes.**

## iv. Box Plot: Summarize the distribution and identify outliers of a numeric variable. (10 Marks)


```python
# Plotting the box plot for 'toss_winner'
plt.figure(figsize=(8, 6))  # Set figure size
plt.boxplot(df_clean['toss_winner'].dropna(), vert=False, patch_artist=True,
            boxprops=dict(facecolor='lightgreen', color='black'),
            flierprops=dict(marker='o', color='red', markersize=8))

# Customizing the plot
plt.title('Box Plot: Toss Winner Distribution', fontsize=14)
plt.xlabel('Toss Winner', fontsize=12)
plt.grid(True)  # Adding grid lines for better readability

# Display the plot
plt.show()
```

### Insights
**The box plot shows the spread of toss winners, with the box representing the interquartile range (IQR), which is the middle 50% of the toss winners' team IDs. The length of the box and the position of the median line within the box indicate the central tendency and the spread of toss winners.**
